# Create an inverted index using Python and execute boolean queries on the same

```python
#importing libraries

import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize
```

```python
#opening the text file 
#tokenization
#specialcharacter removal;

file = open('/content/txt1.txt', encoding='utf8')
file1 = file.read()
file1=file1.split(" ")
file1=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file1])

file = open('/content/txt2.txt', encoding='utf8')
file2 = file.read()
file2=file2.split(" ")
file2=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file2])

file = open('/content/txt3.txt', encoding='utf8')
file3 = file.read()
file3=file3.split(" ")
file3=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file3])


```

```python
#stopword removal
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
  
word_tokens = word_tokenize(file1+" "+file2+" "+file3)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
 
filtered_sentence=" ".join(filtered_sentence) 
```

```python
#selecting unique words from documents and joining them

lst=[set(filtered_sentence.split(" "))]
unique_words=set().union(*lst)
```

```python
#inititalizing a defaultdict which contains words as key and values are 1/2/3
#each key will have 3 values, which implies their presence in these documents 

from collections import defaultdict
docs=defaultdict(list)
for word in unique_words:
    if word!=" " or word!='':
        if word in file1.split(" "):
            docs[word].append('1')
        else :
            pass
        if word in file2.split(" "):
            docs[word].append('2')
        else:
            pass
        if word in file3.split(" "):
            docs[word].append('3')
        else:
            pass
    else:
        continue
```

```python
map_dict={'1':'doc1','2':'doc2','3':'doc3'}
```

```python
#inititalizing a defaultdict which contains words as key and values are 1/2/3
#each key will have 3 values, which implies their presence in these documents

def query_invert(qry):
  for i in qry.split():
    if i=='and':
      qry= qry.replace(i,'&')
    elif i=='or':
      qry=qry.replace(i,'|')
    elif i=='not':
      qry=qry.replace(i,"{'1','2','3'}-")
  for i in qry.split():
    if i.isalpha():
   
      qry=qry.replace(i,repr(set(docs[i])))
  return qry
```

## Evaluation

```python
query='wrote and not foundations'
```

```python
exp=query_invert(query)
final_val=eval(exp)
final_val
```
