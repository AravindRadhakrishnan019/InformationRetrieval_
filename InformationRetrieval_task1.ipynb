# Implement a simple information retrieval model in Python that are capable of processing boolean queries.

```python
#importing libraries

import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize
```

```python
#opening the text file 
#tokenization
#specialcharacter removal;


file = open('/content/txt1.txt', encoding='utf8')
file1 = file.read()
file1=file1.split(" ")
file1=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file1])

file = open('/content/txt2.txt', encoding='utf8')
file2 = file.read()
file2=file2.split(" ")
file2=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file2])

file = open('/content/txt3.txt', encoding='utf8')
file3 = file.read()
file3=file3.split(" ")
file3=" ".join([re.sub('[^A-Za-z0-9]',' ',i) for i in file3])


```

```python
#stopword removal

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
  
word_tokens = word_tokenize(file1+" "+file2+" "+file3)
  
filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
 
filtered_sentence=" ".join(filtered_sentence) 
```

```python
#selecting unique_words from files

lst=[set(filtered_sentence.split(" "))]
unique_words=set().union(*lst)
```

```python
#inititalizing a defaultdict which contains words as key and values are 1/0
#each key will have 3 values, which implies their presence in these documents

from collections import defaultdict
docs=defaultdict(list)
for word in unique_words:
    if word!=" " or word!='':
        if word in file1.split(" "):
            docs[word].append('1')
        else :
            docs[word].append('0')
        if word in file2.split(" "):
            docs[word].append('1')
        else:
            docs[word].append('0')
        if word in file3.split(" "):
            docs[word].append('1')
        else:
            docs[word].append('0')
    else:
        continue
```

```python
docs
```

```python
#function for executing boolean queries

def task1(args):
    vecs={}
    args=args.split(" ")
    for word in args:
        try:
            vecs[word]=int("".join(docs[word]))
        except:
            continue
    # print(vecs)
    exp=[]
    for word in args:
        if word=="and" :
            exp.append("&")
        elif word=="or":
            exp.append("|")
        elif word=="not":
            exp.append("^")
        else:
            exp.append(str(vecs[word]))
    #print(exp)
    res=eval(" ".join(exp))
    #print(res)
        
    return res
```

```python
#mapping function

def doc_matching(res):
  doc_m={1:'doc3',10:'doc2',11:'doc1 and doc2',100:'doc1',110:'doc1 and doc2',111:'doc1 and doc2 and doc3',101:'doc1 and doc3',0:'Not present'}
  return doc_m[res]
```

# Evaluation

```python
res=task1("many and Proscribed and picture")
print(res)
print(doc_matching(res))
```
